```python
import osmnx as ox
import geopandas as gpd
import folium
from shapely.geometry import Point

# Definir la ciudad
ciudad = "Monterrey, Nuevo León, México"
```

```python
# Buscar tiendas de conveniencia
tags = {"shop": "convenience"}
lugares = ox.features_from_place(ciudad, tags=tags)

# Filtrar solo los que tienen nombre 7-Eleven
tiendas_7eleven = lugares[lugares["name"] == "7-Eleven"]

# Verificar cuántas se encontraron
print(f"Total de tiendas 7-Eleven encontradas: {len(tiendas_7eleven)}")
tiendas_7eleven[["name", "geometry"]].head()
```

ANALISIS GEOESPACIAL DE TIENDAS SEVEN ELEVEN

```python
# Centro del mapa
centro_mty = [25.6866, -100.3161]

# Crear mapa
mapa = folium.Map(location=centro_mty, zoom_start=12, tiles="CartoDB positron")

# URL de imagen de 7-Eleven (puedes poner una imagen que esté en línea)
url_icono_7eleven = "https://1000marcas.net/wp-content/uploads/2020/12/7-Eleven-Logo-1968.png"

# Añadir marcadores con el ícono de 7-Eleven
for _, fila in tiendas_7eleven.iterrows():
    if isinstance(fila.geometry, Point):
        lat = fila.geometry.y
        lon = fila.geometry.x
        nombre = fila.get("name", "7-Eleven")

        # Crear marcador con imagen personalizada
        folium.Marker(
            location=[lat, lon],
            popup=nombre,
            icon=folium.CustomIcon(icon_image=url_icono_7eleven, icon_size=(30, 30))  # Icono personalizado
        ).add_to(mapa)

# Mostrar mapa
mapa
```
