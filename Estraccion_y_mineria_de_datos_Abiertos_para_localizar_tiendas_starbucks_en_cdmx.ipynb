```python

# Importar librerías
import osmnx as ox
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from shapely.geometry import Point
from folium.features import CustomIcon
```

```python
# Establecer el área de interés
ciudad = "Ciudad de México, México"
```

```python
# Definir los tags para buscar cafeterías
tags = {"amenity": "cafe"}
```

```python
# Obtener geometrías (cafeterías)
cafeterias = ox.features_from_place(ciudad, tags=tags)
```

```python
# Filtrar solo las tiendas que en el nombre tienen la palabra 'Starbucks'
starbucks_cdmx = cafeterias[cafeterias["name"].str.contains("Starbucks", case=False, na=False)]
```

```python
# Filtrar por cafés que tienen nombre
cafeterias = cafeterias[cafeterias["name"].notna()]
```

```python
# Mostrar cuántas se encontraron
print(f"Total de Starbucks encontrados: {len(starbucks_cdmx)}")
```

```python
# Mostrar las primeras filas
starbucks_cdmx[["name", "geometry"]].head()
```

```python
# Crear un mapa centrado en CDMX
mapa_cdmx = folium.Map(location=[19.4326, -99.1332], zoom_start=12)
```

```python
# Logo desde URL
logo_url = "https://tse4.mm.bing.net/th?id=OIP.Gjbd9dZKJs3o5xHI_KtbbwHaHa&pid=Api&P=0&h=180"
```

```python
# ✅ Esto usa tu logo personalizado
for _, fila in starbucks_cdmx.iterrows():
    if isinstance(fila.geometry, Point):
        lat, lon = fila.geometry.y, fila.geometry.x
        icono_personalizado = CustomIcon(
            logo_url,
            icon_size=(30, 30)
        )
        folium.Marker(
            location=[lat, lon],
            popup=fila["name"],
            icon=icono_personalizado
        ).add_to(mapa_cdmx)


```
